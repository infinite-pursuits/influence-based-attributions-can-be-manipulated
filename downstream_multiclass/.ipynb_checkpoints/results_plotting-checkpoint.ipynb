```python
def get_res(method, eps, targeted_K, dataset, total_iter, num_data, num_seed, lr_list, data_path):
    post_attack_rank = []
    post_acc = [] 
    val_post_attack_rank = []
    val_post_acc = [] 
    for j in range(num_data):
        best_rank = 1000
        best_acc = 0
        val_best_rank = None
        val_best_acc = None
        for seed in range(num_seed):
            for lr in lr_list:
                checkpoint_name_data = f"{data_path}/gd_targeted_attack_multiclass_{dataset}_{1}_{setup}_{eps}_{j}_{lr}_{total_iter}_{seed}.pt"
                try:
#                     cur_targeted_rank_list, val_targeted_rank_list, acc_list, acc_val_list, data_id = torch.load(checkpoint_name_data)
#                     val_targeted_rank_list = [rank.item() for rank in val_targeted_rank_list]
#                     cur_targeted_rank_list = [rank.item() for rank in cur_targeted_rank_list]
# #                     data_id = data_id.item()
#                     torch.save((cur_targeted_rank_list, val_targeted_rank_list, acc_list, acc_val_list, data_id), checkpoint_name_data)
#                     print("success", dataset, j, seed, lr)
                    cur_targeted_rank_list, val_targeted_rank_list, acc_list, acc_val_list, data_id = torch.load(checkpoint_name_data)
                    cur_targeted_rank_list, val_targeted_rank_list = torch.tensor(cur_targeted_rank_list).float(), torch.tensor(val_targeted_rank_list).float()
                except:
                    print("miss", dataset, j, seed, lr)
                    continue
                cur_targeted_rank, acc, val_targeted_rank, val_acc = cur_targeted_rank_list[-1], acc_list[-1], val_targeted_rank_list[-1], acc_val_list[-1]
                if best_rank < targeted_K:
                    if cur_targeted_rank < targeted_K:
                        if acc > best_acc:
                            best_rank, best_acc = cur_targeted_rank, acc
                            val_best_rank, val_best_acc = val_targeted_rank, val_acc
                else:
                    if cur_targeted_rank < best_rank:
                        best_rank, best_acc = cur_targeted_rank, acc
                        val_best_rank, val_best_acc = val_targeted_rank, val_acc
        if val_best_acc is None:
            continue
        post_attack_rank.append(best_rank)
        post_acc.append(best_acc)
        val_post_attack_rank.append(val_best_rank)
        val_post_acc.append(val_best_acc)
    return post_acc, post_attack_rank, val_post_acc, val_post_attack_rank
```

```python
import matplotlib.pyplot as plt

COLOR_S = "#69b3a2"
COLOR_A = "#3399e6"

fontsize = 20

total_iter = 100
num_data = 50
num_seed = 5

eps_list_list = [
    [0.05, 0.1, 0.5],
#     [0.1, 0.5, 1.0],
#     [0.1, 0.5, 1.0],
]

setup_list = [
    "L2_margin", 
#     "L2_margin_uniscal",
#     "L2"
] #, , "closerinit_cos"] #

marker_list = ["o", "v", "s", "p"]

lr_list = [0.01, 0.1]

data_path = "gd_attack_checkpoint_nautilus/gd_attack_checkpoint/"

for dataset in ["cifar10", "pet", "caltech"]:
    for targeted_K in [3, 10]:
        fig, ax1 = plt.subplots(figsize=(9, 6))
        # ax2 = ax1.twinx()

        for setup, marker, eps_list in zip(setup_list, marker_list, eps_list_list):
            success_rate_list = []
            acc_success_list = []
            val_success_rate_list = []
            val_acc_success_list = []
            for eps in eps_list:
                (post_acc, post_attack_rank, val_post_acc, val_post_attack_rank) = get_res(setup, eps, targeted_K, dataset, total_iter, num_data, num_seed, lr_list, data_path)
                post_attack_rank = torch.tensor(post_attack_rank).float()
                success_rate_list.append((post_attack_rank < targeted_K).float().mean())
                post_acc = torch.tensor(post_acc).float()
                acc_success_list.append(post_acc[post_attack_rank < targeted_K].mean())

                val_post_attack_rank = torch.tensor(val_post_attack_rank).float()
                val_success_rate_list.append((val_post_attack_rank < targeted_K).float().mean())
                val_post_acc = torch.tensor(val_post_acc).float()
                val_acc_success_list.append(val_post_acc[val_post_attack_rank < targeted_K].mean())

    #             break
            ax1.plot(success_rate_list, acc_success_list, label=setup+" (val)", marker=marker)
            ax1.plot(val_success_rate_list, val_acc_success_list, label=setup+" (test)", marker=marker)
            for x, y, eps in zip(val_success_rate_list, val_acc_success_list, eps_list):
                ax1.annotate(f"C={eps}", (x, y), fontsize=15)
            for x, y, eps in zip(success_rate_list, acc_success_list, eps_list):
                ax1.annotate(f"C={eps}", (x, y), fontsize=15)
    #             break


        #     ax1.tick_params(axis="y", labelcolor=COLOR_S)
        #     ax2.plot(eps_list, acc_success_list, color=COLOR_A, label=setup, marker=marker)
        #     ax2.tick_params(axis="y", labelcolor=COLOR_A)

        ax1.set_ylabel("model accuracy", fontsize=fontsize)
        # ax2.set_ylabel("accuracy", color=COLOR_A, fontsize=fontsize)

        ax1.set_xlabel("success rate", fontsize=fontsize)
        plt.legend(fontsize=fontsize)
        plt.title(f"Manipulating the targetd data ({dataset}) to top-{targeted_K}", fontsize=fontsize)
        plt.show()
```
