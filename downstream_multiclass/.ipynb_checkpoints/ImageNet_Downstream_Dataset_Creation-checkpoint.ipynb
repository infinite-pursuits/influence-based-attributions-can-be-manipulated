```python
import torchvision
import torchvision.transforms as transforms
# from torchvision.models import resnet50, ResNet50_Weight
from data_utils.cub2011 import Cub2011
import os
from tqdm import tqdm
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
import pytorch_influence_functions as ptif
import pandas as pd
import ast
from sklearn.model_selection import train_test_split
seed = 42
torch.manual_seed(seed)
np.random.seed(seed)
```

```python
def get_dataloader(dataset, tf, data_path="/tmp2/ruihan/", batch_size=16, only_val=False):
    resolution = 224
#     transform_list = [transforms.ToTensor(), 
#                       transforms.Normalize([0.5, 0.5, 0.5], [0.5, 0.5, 0.5]),
#                       CenterCropLongEdge(dataset), 
#                       transforms.Resize(resolution)]
#     tf = transforms.Compose(transform_list)
    if dataset == "bird":
        root = data_path + "CUB_200_2011/"
        train_set = Cub2011(root, train=True, transform=tf)
        test_set = Cub2011(root, train=False, transform=tf)
    elif dataset == "cifar10":
        root = data_path
        train_set = torchvision.datasets.CIFAR10(root, train=True, transform=tf, download=True)
        test_set = torchvision.datasets.CIFAR10(root, train=False, transform=tf, download=True)
    elif dataset == "pet":
        root = data_path
        train_set = torchvision.datasets.OxfordIIITPet(root, split="trainval", transform=tf, download=True)
        test_set = torchvision.datasets.OxfordIIITPet(root, split="test", transform=tf, download=True)
    elif dataset == "car":
        root = data_path
        train_set = torchvision.datasets.StanfordCars(root, split="train", transform=tf)
        test_set = torchvision.datasets.StanfordCars(root, split="test", transform=tf)
    print(type(test_set))
    targets = [target for _, target in test_set]
    #print(targets)
    test_set, val_set = train_test_split(test_set, test_size=0.5, stratify=targets)
    print(type(test_set))
    train_loader = torch.utils.data.DataLoader(train_set, batch_size=batch_size, shuffle=False, num_workers=2)
    val_loader = torch.utils.data.DataLoader(val_set, batch_size=batch_size, shuffle=False, num_workers=2)
    test_loader = torch.utils.data.DataLoader(test_set, batch_size=batch_size, shuffle=False, num_workers=2)
    return train_loader, val_loader, test_loader


def get_feat_label(train_loader, net, checkpoint_name):
    if os.path.exists(checkpoint_name):
        x_feat_tensor, labels_tensor = torch.load(checkpoint_name)
    else:
        x_feat_list = []
        labels_list = []
        for i, data in enumerate(tqdm(train_loader)):
            (images, labels) = data
            x_tf = images.cuda()
            x_feat = net(x_tf)
            x_feat = x_feat.cpu().detach()
            x_feat_list.append(x_feat)
            labels_list.append(labels)
        x_feat_tensor = torch.cat(x_feat_list)
        labels_tensor = torch.cat(labels_list)
        torch.save((x_feat_tensor, labels_tensor), checkpoint_name)
    return x_feat_tensor, labels_tensor
```

```python
pretrained_model = torchvision.models.resnet50(weights=torchvision.models.ResNet50_Weights.IMAGENET1K_V1)
tf = torchvision.models.ResNet50_Weights.IMAGENET1K_V1.transforms()
pretrained_model.eval()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
pretrained_model.to(device)
pretrained_model.fc = torch.nn.Identity()
```

```python
data_path = "/tmp2/ruihan/"
data_path2 = "/tmp2/IF_manipulation_files/"
dataset_list = ["cifar10"] #, "cifar10",  "pet" "car" "bird" doesnt exist
for dataset in dataset_list:
    train_loader, val_loader, test_loader = get_dataloader(dataset, tf, data_path = data_path)
    
    checkpoint_name = f"{data_path2}resnet50_features/{dataset}_train.pt"
    train_x_feat_tensor, train_labels_tensor = get_feat_label(train_loader, pretrained_model, checkpoint_name)
    
    checkpoint_name = f"{data_path2}resnet50_features/{dataset}_val.pt"
    val_x_feat_tensor, val_labels_tensor = get_feat_label(val_loader, pretrained_model, checkpoint_name)
    
    checkpoint_name = f"{data_path2}resnet50_features/{dataset}_test.pt"
    test_x_feat_tensor, test_labels_tensor = get_feat_label(test_loader, pretrained_model, checkpoint_name)
```

```python
train_df = pd.DataFrame({'features': train_x_feat_tensor.tolist(), 'gt_labels': train_labels_tensor.tolist()})
train_df.to_csv(f'{data_path2}resnet50_features/{dataset}_train_features_labels.csv', index=False)

val_df = pd.DataFrame({'features': val_x_feat_tensor.tolist(), 'gt_labels': val_labels_tensor.tolist()})
val_df.to_csv(f'{data_path2}resnet50_features/{dataset}_val_features_labels.csv', index=False)

test_df = pd.DataFrame({'features': test_x_feat_tensor.tolist(), 'gt_labels': test_labels_tensor.tolist()})
test_df.to_csv(f'{data_path2}resnet50_features/{dataset}_test_features_labels.csv', index=False)
```
